```python
##import the library
import os 
import numpy as np
import pandas as pd
import requests
from bs4 import BeautifulSoup
import warnings
warnings.filterwarnings('ignore')
```

```python
##url of website
url = 'https://www.bollywoodhungama.com/celebrity/shah-rukh-khan/box-office/#bh-micro-tab-content'

## making request to get the data
response = requests.get(url)

```

```python
response.content
```

```python
##put the fetched data in beatiful soap
soup =BeautifulSoup(response.content,'html.parser')
```

```python
##renoving the noise from the fetched data
data = soup.find('table', {'id': "example"})
data.findAll('td')
 
```
