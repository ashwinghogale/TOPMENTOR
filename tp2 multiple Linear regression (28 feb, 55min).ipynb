```python
X = df[['R&D Spend', 'Administration', 'Marketing Spend']]
y = df[['Profit']]
y
```

# we create function for multipleregression

```python
def regression(X,y,t):
    X_train,X_test,y_train,y_test = train_test_split(X,y,test_size= t) ## here t is test size | split the data
    model1 = LinearRegression() ## defining the model
    model_training  = model1.fit(X_train,y_train) ## training the model 
    print("training is completed")
    pred = model1.predict(X_test)  ## we are predicting the value
    print(pred)
    accuracy = r2_score(y_test,pred)
   
    return accuracy,model1.coef_,model1.intercept_  ## it show the accuracy
```

```python
Regression(X,y,0.3)
```
